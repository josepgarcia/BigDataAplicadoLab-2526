```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
import pandas as pd
import numpy as np
```

```python
# Crear SparkSession
spark = SparkSession.builder \
    .appName("Ejemplo PySpark") \
    .master("spark://spark-master:7077") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://master:9000") \
    .getOrCreate()

print(f"✅ Spark {spark.version} iniciado")
print(f"✅ Master: {spark.sparkContext.master}")
```

```python
# Crear DataFrame de ejemplo
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
df = spark.createDataFrame(data, ["nombre", "edad"])

df.show()
```

```python
# Guardar en HDFS (módulo1)
df.write.mode("overwrite").parquet("hdfs://master:9000/user/hadoop/ejemplo_spark.parquet")
print("✅ Datos guardados en HDFS")
```

```python
# Leer desde HDFS
df_hdfs = spark.read.parquet("hdfs://master:9000/user/hadoop/ejemplo_spark.parquet")
df_hdfs.show()
print("✅ Datos leídos desde HDFS")
```

```python
# Operaciones con DataFrame
df_hdfs.filter(df_hdfs.edad > 28).show()
```

```python
spark.stop()
```
